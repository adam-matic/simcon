```python
from simcon import *
```

```python
program = """
title Sequence Control, jooss
time 5.0 0.01

t const -3                              # threshold

Jin generator puls 1.0 1.1 1.0          # name op_type start_time end_time amplitude
Js summator t 0.9 Jin 3.0 Ja 3 Sa -3    # name op_type name_of_input1 weight1, name2 weight2 ..
Ja amplifier 0 Js 10 0.01 0 1           # name op_type init name_of_input gain time_const min_lim max_limi

Oin generator puls 2.0 2.1 1.0
Os summator t 0.9 Oin 3.0 Oa 2.0 Ja 1.0 
Oa amplifier 0 Os 10 0.01 0 1

Sin generator puls 3.0 3.1 1.0
Ss summator t 0.9 Sin 3.0 Sa 2.0 Oa 1.0
Sa amplifier 0 Ss 10 0.01 0 1

plot time Ja Oa Sa
"""
a = AnalogComputer(program).run().plot()
```

```python
a.plot(signals=["Jin", "Oin", "Sin"])
```
